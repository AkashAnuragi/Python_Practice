```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matplotlib.__version__
```

```python
data = {
    "Month":['Jan','Feb','Mar','Apr','May','Jun'],
    "Sales":[23445,67533,-67543,45634,37625,76543]
}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.bar(df['Month'], df['Sales'])
```

```python
plt.figure(figsize=(6,3))
plt.bar(df['Month'], df['Sales'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(df['Month'], df['Sales'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie( df['Sales'] )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie( df['Sales'] , labels=df['Month'] )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie( df['Sales'] , labels=df['Month'] , autopct="%0.2f%%" )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot( df['Month'] , df['Sales'])
plt.show()
```

```python
x = np.arange(1,51)
x
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o')
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o', mfc='red')  # mfc-> marker face color
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o', mfc='red' , mec="black") # mec-> marker edges color
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o', mfc='red' , mec="black" , ms=10)   # ms-> marker size
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o', mfc='red' , mec="black" , ms=10 , color='green')
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o', mfc='red' , mec="black" , ms=10 , color='green', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y ,marker='o', mfc='red' , mec="black" , ms=10 , color='green', linestyle='-.')  
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y , marker='o')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot( x, y , marker='o')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(df['Month'], df['Sales'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(df['Month'], df['Sales'] , color='lightgreen')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(df['Month'], df['Sales'] , color=['pink','red','yellow','purple','green','black'])
plt.show()
```

```python
color = []
for sale in df['Sales']:
    if sale>0:
        color.append("lightgreen")
    else:
        color.append('red')
```

```python
plt.figure(figsize=(6,3))
plt.bar(df['Month'], df['Sales'] , color=color)
plt.show()
```
