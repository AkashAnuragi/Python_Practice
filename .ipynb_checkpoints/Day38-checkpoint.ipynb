# OUTLIER
An outlier is a data point or value that is significantly different from the rest of a dataset.
## Example:
Data → 5, 6, 7, 8, 50 → here, 50 is an outlier.

```python
df.describe()
```

# IQR (Interquantile Range)
### Q1 (25 percentile value)
### Q3 (75 percentile value)
### IQR = Q3 - Q1
### min_value = Q1 - 1.5*IQR
### max_value = Q3 + 1.5*IQR

```python
Q1 = df['Gross Sales'].quantile(0.25)
Q1
```

```python
Q3 = df['Gross Sales'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
min_value = Q1 - 1.5*IQR
min_value
```

```python
max_value = Q3 + 1.5*IQR
max_value
```

```python
newdf = df[df['Gross Sales']<max_value]
newdf
```

```python
newdf.shape
```

# Z-Score Method
The Z-score method is used to detect outliers by checking how far a data point is from the mean (average) in terms of standard deviations.
## Z = \frac{X - \mu}{\sigma}


Where:

X = data value
μ = mean of dataset
σ = standard deviation

## Rule for Outliers

Usually:

If Z-score > 3 or Z-score < -3, the value is considered an outlier.

## When to use

Use Z-score method when:

Data is approximately normally distributed
Dataset does not have many extreme values already


```python
df['z_score'] = ( df['Gross Sales'] - df['Gross Sales'].mean() )/df['Gross Sales'].std()
```

```python
df
```

```python
newdf =df[df['z_score']<3]
```

```python
def convertVotes(val):
    if val[-1] == 'M':
        return int(float(val[0:len(val)-1])*1000000)
    elif val[-1] == 'K':        
        return int(float(val[0:len(val)-1])*1000)
    else:
        return int(val)
```

```python
convertVotes('11')
```

```python
df['Votes'] = df['Votes'].map( convertVotes )
```

```python
def convertDuration(val):
    if len(val.split()) == 2:
        val = val.split()
        return int(val[0][0:len(val[0])-1])*60 +  int(val[1][0:len(val[1])-1])
    elif val[-1]=='h':
        return int(val[0:len(val)-1])*60
    elif val[-1]=='m':
        return int(val[0:len(val)-1])
    else:
        return convertDuration( df.Duration.mode()[0] ) 
```

```python
convertDuration('48m')
```

```python
df.Duration = df.Duration.map( convertDuration ) 
```
